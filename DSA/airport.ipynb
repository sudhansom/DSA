```python
def groupAnagrams(words):
    result = {}
    for items in words:
        key = sorted(items)
        key = ''.join(key)
        
        if key in result:
            result[key].append(items)
        else:
            result[key] = [items]
    arr = []
    for key, value in result.items():
        arr.append(value)
    return arr

words = ["yo", "act", "flop", "tac", "foo", "cat", "oy", "olfp"]
print(groupAnagrams(words))
```
