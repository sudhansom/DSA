```python
blocks = [
    {
      "gym": False,
      "school": True,
      "store": False
    },
    {
      "gym": True,
      "school": False,
      "store": False
    },
    {
      "gym": True,
      "school": True,
      "store": False
    },
    {
      "gym": False,
      "school": True,
      "store": False
    },
    {
      "gym": False,
      "school": True,
      "store": True
    }
  ]
g = float('inf')
sc = float('inf')
st = float('inf')
my_dict = {}

for i, item in enumerate(blocks):
    for x, element in enumerate(item):
        j = i
        k = i
        a = 0
        b = 0
        if item[element]:
            continue
        while True:
            if j == 0:
                break
            a += 1
            if blocks[j][element]:
                break
            
            j -= 1
        while True:
            if k == len(blocks)-1:
                break
            b += 1
            if blocks[j][element]:
                break
            
            k += 1
        if x == 0:
            g = min(max(a, b), g)
            a = 0
            b = 0
        if x == 1:
            sc = min(max(a, b), sc)
            a = 0
            b = 0
        if x == 2:
            st = min(max(a, b), st)
            a = 0
            b = 0
    my_dict[i] = max(g, sc, st)
        
print(my_dict)
            
```

```python
blocks = [
    {
      "gym": False,
      "school": True,
      "store": False
    },
    {
      "gym": True,
      "school": False,
      "store": False
    },
    {
      "gym": True,
      "school": True,
      "store": False
    },
    {
      "gym": False,
      "school": True,
      "store": False
    },
    {
      "gym": False,
      "school": True,
      "store": True
    }
  ]
my_dict = {}
for i, item in enumerate(blocks):
    j = 0
    for key, value in item.items():
        if value:
            if i in my_dict:
                my_dict[i].append(1)
            else:
                my_dict[i] = [1]
        else:
            if i in my_dict:
                my_dict[i].append(0)
            else:
                my_dict[i] = [0]
        
    
print(my_dict)
```

```python
blocks = [
    {
      "gym": False,
      "school": True,
      "store": False
    },
    {
      "gym": True,
      "school": False,
      "store": False
    },
    {
      "gym": True,
      "school": True,
      "store": False
    },
    {
      "gym": False,
      "school": True,
      "store": False
    },
    {
      "gym": False,
      "school": True,
      "store": True
    }
  ]
my_dict = {}
for i, block in enumerate(blocks):
    a = [9, 9, 9]
    for x, elem in enumerate(block):
        found_j = False
        found_k = False
        if block[elem]:
            a[x] = 0
            continue
        else:
            j = i
            k = i
            for j in range(i-1, -1, -1):
                if blocks[j][elem]:
                    found_j = True
                    break
            for k in range(i + 1, len(blocks)):
                if blocks[k][elem]:
                    found_k = True
                    break
        if i - j and k - i and found_j and found_k:
            a[x] = min(i-j, k - i)
        elif i - j and found_j:
            a[x] = i - j
        elif k - i and found_k:
            a[x] = k - i
#         print(i, i - j, k - i)
    my_dict[i] = max(a)
    
    
print(my_dict)
        
```
