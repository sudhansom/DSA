```python
class Bst:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
    def add_child(self, data):
        if self.data == data:
            return
        if data < self.data:
            if self.left:
                self.left.add_child(data)
            else:
                self.left = Bst(data)
        if data > self.data:
            if self.right:
                self.right.add_child(data)
            else:
                self.right = Bst(data)
    def in_order_traversal(self):
            elements = []
            if self.left:
                elements += self.left.in_order_traversal()
            elements.append(self.data)
            if self.right:
                elements += self.right.in_order_traversal()
            return elements
    def depth_first_search(self):
        print(self.data)
        if self.left:
            self.left.depth_first_search()
        if self.right:
            self.right.depth_first_search()
    def find_closest_value(self,  target, closest='', diff=100):
        if self.data == target:
            return self.data
        if abs(self.data - target) < diff:
            diff = abs(self.data - target)
            closest = self.data
        if target < self.data:
            if self.left:
                closest = self.left.find_closest_value(target, closest, diff)
        if target > self.data:
            if self.right:
                closest = self.right.find_closest_value(target, closest, diff)
        return closest
    def branch_sum(self, sums=0, arr=[]):
        sums += self.data
        if self.left is None and self.right is None:
            arr.append(sums)
        if self.left:
            self.left.branch_sum(sums, arr)
        if self.right:
            self.right.branch_sum(sums, arr)
        return arr
    def max_node_depth(self, cnt=-1, arr=[]):
        cnt += 1
        if self.left is None and self.right is None:
            arr.append(cnt)
        if self.left:
            self.left.max_node_depth(cnt, arr)
        if self.right:
            self.right.max_node_depth(cnt, arr)
        return arr
    def sum_of_node_depth(self, cnt=-1, arr=[]):
        cnt += 1
        arr.append(cnt)
        if self.left:
            self.left.sum_of_node_depth(cnt, arr)
        if self.right:
            self.right.sum_of_node_depth(cnt, arr)
        return sum(arr)
    def sum_of_node(self, cnt=-1, sums=0):
        cnt += 1
        sums += cnt
        if self.left:
            sums = self.left.sum_of_node(cnt, sums)
        if self.right:
            sums = self.right.sum_of_node(cnt, sums)
        return sums
    def no_of_nodes(self, total=0):
        total += 1
        if self.left:
            total = self.left.no_of_nodes(total)
        if self.right:
            total = self.right.no_of_nodes(total)
        return total
    
    def invert_tree(self): # recursive approach
        self.left, self.right = self.right, self.left
        if self.left:
            self.left.invert_tree()
        if self.right:
            self.right.invert_tree()
        
    def invert_tree_iterative(self): # iterative method
        queue = [self]
        while queue:
            current = queue.pop(0)
            if current:
                current.left, current.right = current.right, current.left
                queue.append(current.left)
                queue.append(current.right)
                
    def depth_first_search(self, arr = []):
        arr.append(self.data)
        if self.left:
            self.left.depth_first_search(arr)
        if self.right:
            self.right.depth_first_search(arr)
            
        return arr
    
    def breadth_first_search(self, arr = []):
        queue = [self]
        while queue:
            current = queue.pop(0)
            arr.append(current.data)
            if current.left:
                queue.append(current.left)
            if current.right:
                queue.append(current.right)
            
        return arr
        
        
    
```

```python
def nonConstructibleChange(arr):
    for i in range(1, arr[-1]):
        if check(i, arr):
            continue
        else:
            return i
    return True

def check(i, arr):
    if i in arr:
        return True
    else:
        change = i - 1
        for elem in arr:
            if elem + change == i:
                return True
arr = [1, 1, 2, 3, 5, 7, 22]
print(nonConstructibleChange(arr))
```
