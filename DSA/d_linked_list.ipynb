```python
class Node:
    def __init__(self, data, prev=None, next=None):
        self.data = data
        self.prev = prev
        self.next = next
        
class D_link_list:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0
        
    def append(self, data):
        self.length += 1
        if self.head is None:
            self.head = Node(data)
            self.tail = self.head
            return
        itr = self.head
        while itr.next:
            itr = itr.next
        itr.next = Node(data,itr, None)
        self.tail = itr.next
        return True
        
    def show_all_r(self):
        print(self.head.next.data)
        print(self.tail.prev.data)
        if self.head is None:
            print("empty list....")
            return
        itr = self.tail
        print(None, end='<--->')
        while itr:
            print(itr.data, end="<--->")
            itr = itr.prev
        print(None)
        
    def show_all(self):
        print(self.head.next.data)
        print(self.tail.prev.data)
        if self.head is None:
            print("empty list....")
            return
        itr = self.head
        print(None, end='<--->')
        while itr:
            print(itr.data, end="<--->")
            itr = itr.next
        print(None)
        
    def __delitem__(self, indx):
        if self.head is None:
            print("empty list")
            return
        if indx == 0:
            self.head = self.head.next
            self.head.prev = None
            return
        if indx == self.length - 1:
            self.tail = self.tail.prev
            self.tail.next = None
            return
        cnt = 0
        itr = self.head
        while itr:
            if cnt == indx - 1:
                itr.next = itr.next.next
                itr.next.prev = itr
                return
            itr = itr.next
            cnt += 1
            
    def insert(self, indx, data):
        if indx == 0:
            self.head = Node(data, None, self.head)
            self.head.next.prev = self.head
            return
        if indx >= self.length:
            self.tail = Node(data, self.tail, None)
            self.tail.prev.next = self.tail
            return
        cnt = 0
        itr = self.head
        while itr:
            if cnt == indx - 1:
                itr.next = Node(data, itr, itr.next)
                itr.next.next.prev = itr.next
                return
            cnt += 1
            itr = itr.next
            
    def __setitem__(self, indx, data):
        if indx == 0:
            self.head.data = data
            return
        if indx >= self.length:
            print('index out of range')
        itr = self.head
        cnt = 0
        while itr:
            if indx == cnt:
                itr.data = data
                return
            cnt += 1
            itr = itr.next
            
    def __getitem__(self, indx):
        itr = self.head
        cnt = 0
        while itr:
            if indx == cnt:
                return itr.data
            itr = itr.next
            cnt += 1
            
    def __len__(self):
        return self.length
    
    def __append__(self):
        print('kldfksd')
        return 'yess'
    
    
dll = D_link_list()
dll.append(0)
dll.append(1)
dll.append(2)
dll.append(3)
dll.append(4)
dll.append(5)
dll.append(6)
#  dll.delete(1)
# dll.update(6, 333)
dll[6] = 8888
del dll[6]
dll.show_all()
dll.show_all_r()
dll[3]
len(dll)


            
        
```
